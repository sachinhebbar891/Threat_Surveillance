```python
import base64
import os
from langchain_core.messages import HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
# if "GOOGLE_API_KEY" not in os.environ:
#     os.environ["GOOGLE_API_KEY"] = "Enter API KEY here"
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash-001")
```

```python
path = "/content/videos/"
all_videos = os.listdir(path)
all_videos
```

```python
video_responses = {}
video_mime_type = "video/mp4"
for video in all_videos:
    video_file_path = path + video
    with open(video_file_path, "rb") as video_file:
        encoded_video = base64.b64encode(video_file.read()).decode("utf-8")

    message = HumanMessage(
        content=[
            {"type": "text", \

            "text": """
            You are an AI assistant working in a bank's alarm monitoring center.
            You will receive a video input whenever there is an alarm in the bank or any other center.

            Your goal is to review the video input sent to you and do the following 2 activities:
            1. Identify the alarm type.
              - If the video is showing some criminal activity then its a true alarm.
              - If the video is showing a mundane or non-criminal activity then its a false positive.
            2. Provide an alarm summary explaining what's happening in the video and why that makes it a true alarm or false positive.

            Be precise in your responses.

            Provide response in the following json format:
            {
              Alarm type: True alarm/False positive,
              Alarm Summary: Summarize why you find the video to be a true alarm/false positive
            }
            """

            },
            {
                "type": "media",
                "data": encoded_video,
                "mime_type": video_mime_type,
                "response_mime_type": "application/json",
            },
        ]
    )

    response = llm.invoke([message])

    video_responses[video] = response.content
```

```python
with open('/content/output.txt', 'w') as file:
    file.write(str(video_responses))
```
