```python
import yt_dlp

def download_youtube_video(youtube_url, output_path='./', filename='video.mp4'):
    ydl_opts = {
        'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]',
        'outtmpl': f'{output_path}/{filename}',
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([youtube_url])
```

```python
download_youtube_video(
    youtube_url="https://www.youtube.com/watch?v=wWEHimK-bNw",
    output_path="/content/",
    filename="atm_robbery.mp4"
)
```
